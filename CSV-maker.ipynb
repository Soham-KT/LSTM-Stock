```python
stock_train = []
stock_test = []

for comp in nifty50_comapanies:
    stock_train.append(yf.download(comp, start='2014-01-01', end='2022-12-31'))
    stock_test.append(yf.download(comp, start='2023-01-01', end='2023-12-31'))
    
```

```python
for i in range(len(nifty50_comapanies)):
    df_train = pd.DataFrame(stock_train[i])
    df_test = pd.DataFrame(stock_test[i])
    
    df_train.insert(0, 'Date', df_train.index, True)
    df_test.insert(0, 'Date', df_test.index, True)
    
    df_train.reset_index(drop=True, inplace=True)
    df_test.reset_index(drop=True, inplace=True)
    
    df_train.to_csv('Data/' + nifty50_comapanies[i] + '-train.csv', index=False)
    df_test.to_csv('Data/' + nifty50_comapanies[i] + '-test.csv', index=False)
```
